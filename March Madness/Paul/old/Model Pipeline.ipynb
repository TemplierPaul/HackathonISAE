## Dataset Creation

```python
seasons = list(MTSeasons_Stage1['Season'].unique())
team_ids = list(MTeams_Stage1['TeamID'])
```

```python
team_stats = []

for s in seasons:
    df = MRegularSeasonCompactResults_Stage1[MRegularSeasonCompactResults_Stage1['Season']==s]
    print('Season %d: %d matches\t\t'%(s, len(df)), end='\r')

    for t in team_ids:
        w = len(df[df['WTeamID']==t])
        l = len(df[df['LTeamID']==t])
        p = w + l
        if p==0:
            wr = 0
        else:
            wr = w / p
        d = {
            'Season':s,
            'TeamID':t,
            'W':w,
            'L':l,
            'Played':p,
            'Win rate':wr
        }
        team_stats.append(d)
```

```python
team_stats_df = pd.DataFrame(team_stats)
team_stats_df = team_stats_df[['Season', 'TeamID', 'Played',  'W', 'L', 'Win rate']]
team_stats_df.head()
```

```python
team_stats_df.to_csv('stats_v1.csv', columns=['Season', 'TeamID', 'Played',  'W', 'L', 'Win rate'])
```

## Data aggregation
### Match history

```python
m = MNCAATourneyCompactResults_Stage1

m1 = pd.DataFrame()
m1['Season'] = m['Season']
m1['ID_Team1'] = m['WTeamID']
m1['ID_Team2'] = m['LTeamID']
m1['Team1_Home'] = (m['WLoc']=='H').astype(int)
m1['Team2_Home'] = (m['WLoc']=='A').astype(int)
m1['Team1_Win'] = 1

m2 = pd.DataFrame()
m2['Season'] = m['Season']
m2['ID_Team1'] = m['LTeamID']
m2['ID_Team2'] = m['WTeamID']
m2['Team1_Home'] = (m['WLoc']=='A').astype(int)
m2['Team2_Home'] = (m['WLoc']=='H').astype(int)
m2['Team1_Win'] = 0

match_histo = pd.concat([m1, m2])
results = match_histo['Team1_Win']
match_histo = match_histo.drop(columns = ['Team1_Win'])
match_histo.head()
```

### Data about each team, each season

```python
path = 'stats_v1.csv'
data_teams = pd.read_csv(path, index_col=0)
data_teams.head()
```

### Merge Training dataset

```python
full_data = pd.merge(match_histo, data_teams, how='left', left_on=['Season', 'ID_Team1'], right_on=['Season', 'TeamID'])
full_data = pd.merge(full_data, data_teams, how='left', left_on=['Season', 'ID_Team2'], right_on=['Season', 'TeamID'], suffixes=['_Team1', '_Team2'])
full_data.head(10)
```

## Test set generation
We generate a set of every possible combination of teams to assess victory probabilities

```python
season = 2019

f = MNCAATourneySeeds_Stage1['Season']==2018
mad_teams = list(MNCAATourneySeeds_Stage1[f]['TeamID'])

generated_matches = []
duos = []
for i in mad_teams:
    for j in mad_teams:
        if (j, i) not in duos:
            duos.append((i, j))
            if i != j:
                d = {
                    'Season':season,
                    'ID_Team1':i,
                    'ID_Team2':j,
                    'Team1_Home':0,
                    'Team2_Home':0
                }
                generated_matches.append(d)
```

```python
generated_matches = pd.DataFrame(generated_matches)
generated_matches.head()
```

### Merge evaluation dataset

```python
to_predict = pd.merge(generated_matches, data_teams, how='left', left_on=['Season', 'ID_Team1'], right_on=['Season', 'TeamID'])
to_predict = pd.merge(to_predict, data_teams, how='left', left_on=['Season', 'ID_Team2'], right_on=['Season', 'TeamID'], suffixes=['_Team1', '_Team2'])
to_predict.head(10)
```

## Model
### Train

```python
model = RandomForestClassifier(n_estimators = 10)
model.fit(full_data, results)
```

### Evaluate

```python
team1_proba = model.predict_proba(to_predict)
```

```python
result = pd.DataFrame()
result[['Année', 'ID_Team1', "ID_Team2"]] = to_predict[['Season', 'ID_Team1', "ID_Team2"]]
result['Predic_Team1'] = team1_proba[:,0]
result.head()
```
